## Histograms

```python
# Load some images, calculate and plot their color histograms, one color at the time
# try it with different color spaces BGR, HSV, etc

path = './img/'
rick  = cv2.imread(path + 'rick.jpg')
day = cv2.imread(path + "day1.jpg")
```

[Hint](# "cv2.calcHist") 


```python
# Now try to plot all 3 channels at the same time, does the histogram give you an idea of what the image looks like?
fig, axes = plt.subplots(1, 3, figsize=(15, 4))



rick_b_h = cv2.calcHist([rick], [0], None, [255], [0,255])
axes[0].plot(rick_b_h, color='blue')


rick_g_h = cv2.calcHist([rick], [1], None, [255], [0,255])
axes[1].plot(rick_g_h, color='green')

rick_r_h = cv2.calcHist([rick], [2], None, [255], [0,255])
axes[2].plot(rick_r_h, color='red')
```

```python
colors = ['b', 'g', 'r']
for i, color in enumerate(colors):
    histogram = cv2.calcHist([rick],[i],None,[255],[0,255])
    plt.plot(histogram,color=color)
    plt.xlim(0,256)
plt.show()
```
