```python
gray_chess = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect the corners using the cornerHarris

corners = cv2.cornerHarris(gray_chess, 15, 7, 15)

# Display the image with the detected corners 

plt.imshow(corners, cmap='gray')
```

```python
img = cv2.imread('img/chess.jpg')

# Convert img to grayscale

gray_chess = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


# Detect the corners using goodFeaturesToTrack

max_corners = 100
quality = 0.01
distance = 10

corners = cv2.goodFeaturesToTrack(gray_chess, max_corners, quality, distance)

# Draw the corners on top of img and show it on screen (make sure the corners are visible!)

for corner in corners:
    x, y = corner[0]
    x = int(x)
    y = int(y)
    cv2.rectangle(gray_chess, (x-5, y-5), (x+5, y+5), (0, 255, 0), -1)

plt.figure(figsize=(7,7))    
plt.imshow(gray_chess, cmap='gray')
```
