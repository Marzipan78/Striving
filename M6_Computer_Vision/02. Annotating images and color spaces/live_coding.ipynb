```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def imgshow(img):
    plt.figure(figsize=(10,10))
    plt.imshow((img[:,:,::-1]))
```

```python
path = './img/'
```

```python
rick = cv2.imread(path + 'rick.jpg')
rick_copy = rick.copy()
```

```python
imgshow(rick_copy)
```

```python
hist_blue = cv2.calcHist([rick_copy],[0],None,[255],[0,255])
plt.plot(hist_blue,color='b')
```

```python
hist_green = cv2.calcHist([rick_copy],[1],None,[255],[0,255])
plt.plot(hist_green,color='g')
```

```python
hist_red = cv2.calcHist([rick_copy],[2],None,[255],[0,255])
plt.plot(hist_red,color='r')
```

```python
colors = ['b', 'g', 'r']
for i, color in enumerate(colors):
    histogram = cv2.calcHist([rick_copy],[i],None,[255],[0,255])
    plt.plot(histogram,color=color)
    plt.xlim(0,256)
plt.show()
```

```python
trans_mat = np.float32([[1,0,50],
                        [0,1,100]])
```

```python
trans_mat
```

```python
h, w = rick_copy.shape[:2]
h
```

```python
t_img = cv2.warpAffine(rick_copy,trans_mat,(w,h))
```

```python
imgshow(t_img)
```

```python
x_center = w/2
y_center = h/2
rot_mat = cv2.getRotationMatrix2D((x_center,y_center),50,0.5)
```

```python
rot_mat
```

```python
rot_img =cv2.warpAffine(rick_copy,rot_mat,(w,h))
imgshow(rot_img)
```
