```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python

def imgshow(img):
    plt.figure(figsize=(10,10))
    plt.imshow((img[:,:,::-1]))
```

```python
path = './img/'

```

```python
g1 = cv2.imread(path + 'jon1.png')
g1_copy = g1.copy()
```

```python
imgshow(g1_copy)
```

```python
start_point1 = (400, 10)
start_point2 = (400, 30)  
# Ending coordinate, here (125, 80)
# represents the bottom right corner of rectangle
end_point1 = (410, 20)
end_point2 = (410, 40)
# Black color in BGR
color = (255, 0, 0)
color2 = (20, 100, 255)
# Line thickness of -1 px
# Thickness of -1 will fill the entire shape
thickness = -1
   
# Using cv2.rectangle() method

# Draw a rectangle of black color of thickness -1 px
g1_colored = cv2.rectangle(g1_copy, start_point1, end_point1, color, thickness)
cv2.putText(g1_colored , 'Train Loss', (413,20), cv2.FONT_HERSHEY_COMPLEX, .5, (0,0,0),1, cv2.LINE_AA)
cv2.putText(g1_colored , 'Validation Loss', (413,40), cv2.FONT_HERSHEY_COMPLEX, .5, (0,0,0),1, cv2.LINE_AA)
g1_coloreds = cv2.rectangle(g1_colored, start_point2, end_point2, color2, thickness)
imgshow(g1_coloreds)
```

```python
cv2.imwrite('loss_graph.jpg', g1_coloreds)
```
