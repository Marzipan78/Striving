### To edit an image making a copy of it is always a good idea!

```python
# Load an image
# And make a copy of the original image to avoid manipulating the original image
path = './img/'

rick = cv2.imread(path + 'rick.jpg')
rick_copy = rick.copy()
```

[Hint](# "Numpy .copy function") 

```python

def imgshow(img):
    plt.figure(figsize=(10,10))
    plt.imshow((img[:,:,::-1]))
```

```python
imgshow(rick_copy)
```

## Drawing

### Drawing lines

```python
# Draw some lines on top of the image 
cv2.line(rick_copy,(350,100),(850,100),(255,0,0),4)
imgshow(rick_copy)
```

[Hint](# "cv2.line function") 

### Circles

```python
# Draw some circles on top of the image
cv2.circle(rick_copy,(580,470),5,(0,0,0),-1)
cv2.circle(rick_copy,(700,445),5,(0,0,0),-1)
cv2.circle(rick_copy,(1240,590),5,(0,0,0),-1)
cv2.circle(rick_copy,(1360,590),5,(0,0,0),-1)
imgshow(rick_copy)
```

[Hint](# "cv2.circle") 

### Ellipses

```python
# Draw some elipses on top of the image
angle = 0
  
startAngle = 0
  
endAngle = 360
   
# Blue color in BGR
color = (255, 0, 0)
   
# Line thickness of -1 px
thickness = 3
cv2.ellipse(rick_copy, (1300,375), (150, 50),angle,startAngle,endAngle,color,thickness)
imgshow(rick_copy)
```

[Hint](# "cv2.ellipse")
