```python
import pandas as pd
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
df = pd.read_csv('../transcript.csv')
```

```python
test_df = df.iloc[0:5]
```

```python
docs = list(df["text"])
#docs = [nlp(doc) for doc in docs]
docs
```

```python
def listToString(s): 
    
    # initialize an empty string
    str1 = " " 
    
    # return string  
    return (str1.join(s))

print(listToString(docs))
```

```python
doc = nlp(listToString(docs))
doc
```

```python
for token in doc:
    print(token)
```

```python
from spacy import displacy

displacy.render(doc, style='ent')
```

```python
def show_ents(doc):
    if doc.ents:
        for ent in doc.ents:
            print(ent.text,  ent.label_)
    else:
        print("No entities to show")
```

```python
doc = nlp(listToString(docs))
```

```python
show_ents(doc)
```
