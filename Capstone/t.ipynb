```python
from youtube_transcript_api import YouTubeTranscriptApi
from youtube_transcript_api.formatters import TextFormatter
from youtube_transcript_api.formatters import WebVTTFormatter

```

```python
video_id = "c0zLzSgBO8o"
```

```python
transcript = YouTubeTranscriptApi.get_transcript(video_id=video_id)
```

```python
transcript
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(transcript)
```

```python
df
```

```python
df['start'] = df['start'].astype(int)
```

```python
from time import strftime
from time import gmtime
```

```python

strftime("%H:%M:%S", gmtime(df['start'].iloc[0]))
```

```python
timestamp = []
for i in range(0,len(df)):
    timestamp.append(strftime("%H:%M:%S", gmtime(df['start'].iloc[i])))
```

```python
duration = []
for i in range(0,len(df)):
    duration.append(strftime("%H:%M:%S", gmtime(df['duration'].iloc[i])))
```

```python
timestamp
```

```python
df['timestamp'] = timestamp
df['duration_'] = duration
```

```python
df
```

```python
df.to_csv('transcript.csv', index=False)
```
