```python
from requests_html import HTMLSession
```

```python
def getTitle(url):
    s = HTMLSession()
    r = s.get(url)
    r.html.render(sleep=1)

    book = {
            'title': r.html.xpath('//*[@id="all_votes"]/table/tbody/tr[1]/td[3]/a/span' , first=True).text

    }
    print(book)
    return book
getTitle('https://www.goodreads.com/list/show/264.Books_That_Everyone_Should_Read_At_Least_Once?page=1#')
```
