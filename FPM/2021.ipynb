## Q2

### ABC stock is currently trading at $150. In the next period, the price will either go up by 15% or down by
### 7%. The risk-neutral probability of going up (q) is equal to 0.73. You also have a riskfree asset available
### (riskless deposit). What is the fair forward price of ABC.
### Answer: You have a stock with known payoffs and prices, and you have a riskless deposit with unknown
### rate. Recall that risk-neutral expectation of the stock payoffs is equal to the forward rate:

#### Su = S * (1 + ru / 100)
#### Sd = S * (1 - rd / 100)
#### F = q*Su + (1 - q) * Sd

```python
S, ru, rd, q = 150, 15, 7, 0.73
Su = S * (1 + ru / 100)
Sd = S * (1 - rd / 100)
F = q * Su + ( 1 - q ) * Sd
F
```

# Q7 
### A stock price is currently $29. It  is  known  that  at  the  end  of  two  months  it will be either $22 or $33. The risk-free interest rate is 7% per annum with continuous compounding. Suppose ST is the stock price at the end of two months. What is the price of a derivative that pays off ST*ST (that is, ST squared)?

#### Answer: We can use risk-neutral approach:

#### ğ‘† âˆ— ğ‘’ğ‘¥ğ‘(ğ‘Ÿğ‘“/100/6) = ğ‘ âˆ— ğ‘†ğ‘¢ + (1 âˆ’ ğ‘) âˆ— ğ‘†ğ‘‘
#### ğ‘ = (ğ‘† âˆ— ğ‘’ğ‘¥ğ‘(ğ‘Ÿğ‘“/6) âˆ’ ğ‘†ğ‘‘)/(ğ‘†ğ‘¢ âˆ’ ğ‘†ğ‘‘)
#### ğ· = (ğ‘ âˆ— ğ‘†ğ‘¢ + (1 âˆ’ ğ‘) âˆ— ğ‘† ) âˆ— ğ‘’ğ‘¥ğ‘(âˆ’ğ‘Ÿğ‘“/100/6)

```python
S,Sd,Su,rf = 29,22,33,7

q = (S*np.exp(rf/100/6) - Sd)/(Su-Sd)
Du, Dd = Su**2, Sd**2
D = (q*Du + (1-q)*Dd)*np.exp(-rf/100/6)
D
```

#### Suppose you have a stock that pays either $114 or $90 in the next period and a call option on the stock with strike $105 and current price $10. In addition, there is a risk-free asset that pays 12% per period. What is the current stock price?
#### Answer:
#### We can use either replicating portfolio or risk-neutral approach. Let us use the latter.
#### From the risk-neutral pricing we know:
#### ğ¶ğ‘¢ = ğ‘šğ‘ğ‘¥(ğ‘ 1 âˆ’ ğ¾, 0)
#### ğ¶ğ‘‘ = ğ‘šğ‘ğ‘¥(ğ‘ 2 âˆ’ ğ¾, 0)
#### ğ‘ âˆ— ğ¶ğ‘¢ + (1 âˆ’ ğ‘) âˆ— ğ¶ğ‘‘ = ğ¶0 âˆ— (1 + ğ‘Ÿğ‘“/100)
#### ğ‘ = (ğ¶ âˆ— (1 + ğ‘Ÿğ‘“/100) âˆ’ ğ¶ğ‘‘)/(ğ¶ğ‘¢ âˆ’ ğ¶ğ‘‘)
#### ğ‘† = (ğ‘ âˆ— ğ‘ 1 + (1 âˆ’ ğ‘) âˆ— ğ‘ 2)/(1 + ğ‘Ÿğ‘“/100

```python
s1,s2,K,C0,rf = 114,90,105,10,12
Cu = max(s1-K,0)
Cd = max(s2-K,0)
# $q*Cu + (1-q)*Cd = C*(1+rf/100)$
q = (C0*(1+rf/100) - Cd)/(Cu-Cd)
S = (q*s1 + (1-q)*s2)/(1+rf/100)
S
```
