### Problem 9 (10 points/ computations)
#### A short forward on silver for delivery in twelve months is entered into at the fair price of per troy ounce (oz t). The volume of the contract is 10,000 oz t. Storing silver for a three-month period is charged 0.02 $/ozt at the beginning of the period. Interest rates are 6% per annum, quarterly compounded. 1. Find the spot price of silver and the value of your forward contract at inception. 2. Six months later, the spot price of silver has gone down to $12 per troy ounce, and the interest rate is still 6â§µ% p.a. What are the current forward price for the same delivery date as before and the value of the forward position entered into six months before?

```python
# 1. The value of the forward contract at inception is 0 
# The spot price of silver S0 can be backed out of the forward price F012
# PV(F012) = S0 + PV(Holding Costs) - PV(Benefits)
r012 = 0.06 # rate
freq = 4 # compounding frequency
F012 = 14 # current forward
HC = 0.02 # holding costs per period of time
S6 = 12 # asset spot price in the future 
# PV of the forward 
df = 1/(1+r012/freq)
PV_F012 = F012*df**4
# PV of holding costs
PV_HC = HC*(1+df+df**2+df**3)
# compute S0
S0 = PV_F012 - PV_HC
print('1. Current value is ',0)
print(' Current spot is ', round(S0,2))
# 2. Compute new forwrd price, future PNL, discount to today to get the value of PNL
# PV(F612) = S6 + PV(Holding Costs) - PV(Benefits)
PV_HC = HC*(1+df) # just two payments -- now and in 3 months 
PV_F612 = S6 + PV_HC # PV of forward
F612 = PV_F612/df**2 # compound for two periods 
# future PNL discounted for two periods 
Notional = 10000
Pos_VAL = - Notional * (F612 - F012)*df**2 # note that it's the SHORT forward posi
print('2. Current forward is ', round(F612,2))
print(' Position value ', round(Pos_VAL,2))
```

### Problem 15 (10 points / computations)
#### Stock A is traded at 100 now, and pays dividend of 10 in six months from now. In one year the realization of the price can be either 125 with probability 75%, or 80 with probability 25%, and the riskfree rate is 5% p.a. with semi-annual compounding. The stock can be traded only now and in one year; money-market account can be traded now, in six months and in one year.
1. Find the forward price with maturity 1 year
2. Using replicating portoflio approach with underlying stock and money-market account, find the price of
a call option with strike K = 100 expiring in one year. Note that the when the option is exercised, the exdividend stock is delivered to the option holder, that is, when you exercise an option, you do not get
dividends

```python
# parameters 
S0 = 100
Su = 125
Sd = 80
K = 100
Cu = max(0,Su-K)
Cd = max(0,Sd-K)
Div = 10
r = 0.05
freq= 2
mm = (1+r/freq) # money-market account growth per period
#1. ####### forward
# use PV(F) = S0 - PV(Holding Benefits)
PV_F012 = S0 - Div/mm
F012 = PV_F012*mm**2
print('Forward price is ', round(F012,2))
#2. ####### replicating portfolio 
# note that if we hold a stock, it will pay us dividends in six months, and
# we will earn interest in this amount foro six more months; thus, we need to take 
# it into account when we construct the portoflio
# write down two equations, one for each future state
# the payoffs from the option are + 25 and 0 in up and down states, respectively
# Delta*Su + Delta*Div*mm + B*mm**2 = Cu
# Delta*Sd + Delta*Div*mm + B*mm**2 = Cd
Delta = (Cu-Cd)/(Su-Sd)
B = (Cd - Delta*(Sd + Div*mm))/mm**2
print('Delta:', round(Delta,2), ' B', round(B,2))
# find option price now 
C0 = Delta*S0 + B
print('Call price is ', round(C0,2))
######## EXTRA
# we can also solve it using linear equations system: 
X = np.array([[Su+Div*mm,Sd+Div*mm],[mm**2,mm**2]]) # payoff matrix 
Payoff = np.array([Cu,Cd]).reshape((2,1))
Ptf = np.linalg.inv(X.T).dot(Payoff)
Prices = [S0,1]
C0syst = Ptf.T.dot(np.array(Prices))[0]
print('Call price (solving system) is ', round(C0,2))
```

### Problem 18 (6 points/ computation)
#### Bond W with a face value of 100 has a coupon of 10% (semiannual basis) and a maturity of one year, and is trading at 98.4496. If the one-year spot rate is 12%, what is the six month rate? All rates are with semiannual compounding.

```python
FV = 100
P0 = 98.4496
freq = 2
C = 0.10/freq*FV
r12 = 0.12
# start from the price 
# P0 = C / (1+r6/2) + (FV+C)/(1+r12/2)**2
# C / (1+r6/2) = P0 - (FV+C)/(1+r12/2)**2
# r6 = (C/ (P0 - (FV+C)/(1+r12/2)**2) - 1)*2
r6 = (C/(P0 - (FV+C)/(1+r12/2)**2) - 1)*2
print('Six-month spot rate is ', round(r6*100,2), '% p.a.')
```
