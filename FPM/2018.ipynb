### Problem 9 (10 points/ computations)
#### A short forward on silver for delivery in twelve months is entered into at the fair price of per troy ounce (oz t). The volume of the contract is 10,000 oz t. Storing silver for a three-month period is charged 0.02 $/ozt at the beginning of the period. Interest rates are 6% per annum, quarterly compounded. 1. Find the spot price of silver and the value of your forward contract at inception. 2. Six months later, the spot price of silver has gone down to $12 per troy ounce, and the interest rate is still 6â§µ% p.a. What are the current forward price for the same delivery date as before and the value of the forward position entered into six months before?

```python
# 1. The value of the forward contract at inception is 0 
# The spot price of silver S0 can be backed out of the forward price F012
# PV(F012) = S0 + PV(Holding Costs) - PV(Benefits)
r012 = 0.06 # rate
freq = 4 # compounding frequency
F012 = 14 # current forward
HC = 0.02 # holding costs per period of time
S6 = 12 # asset spot price in the future 
# PV of the forward 
df = 1/(1+r012/freq)
PV_F012 = F012*df**4
# PV of holding costs
PV_HC = HC*(1+df+df**2+df**3)
# compute S0
S0 = PV_F012 - PV_HC
print('1. Current value is ',0)
print(' Current spot is ', round(S0,2))
# 2. Compute new forwrd price, future PNL, discount to today to get the value of PNL
# PV(F612) = S6 + PV(Holding Costs) - PV(Benefits)
PV_HC = HC*(1+df) # just two payments -- now and in 3 months 
PV_F612 = S6 + PV_HC # PV of forward
F612 = PV_F612/df**2 # compound for two periods 
# future PNL discounted for two periods 
Notional = 10000
Pos_VAL = - Notional * (F612 - F012)*df**2 # note that it's the SHORT forward posi
print('2. Current forward is ', round(F612,2))
print(' Position value ', round(Pos_VAL,2))
```

#### Problem 10 (7 points/ computations)
#### Consider a forward contract on a bond. Suppose that: Spot price: S = 100; contract length: T = 12 month Coupon of will be paid to bond holders in 6 months and coupon of 6 in 12 months Term structure is not flat, and the discount functions for 6 and 12 months are 0.975 and 0.945, respectively.
1. Find the present value of holding benefits for the bond
2. Find the fair forward price

```python
# inputs 
S = 100
t = 12/12
C = np.array([4,6]) # cash flows 
DF = np.array([0.975,0.945]) # discount functon
# PV of holding benefits 
PV_HB = (C*DF).sum() # multiply coupons by discount function and add up 
F012 = (S - PV_HB)/DF[1] # compound the (S - PV_HB)
print('1. PV of holding benefits ', round(PV_HB,2))
print('2. Forward price ', round(F012,2))
```
