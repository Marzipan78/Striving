# Enumerate vs Range

```python
for i in range(len(my_list)):
    print(i, "-->", my_list[i])
```

```python
for i,item in enumerate(my_list):
    print(i, "-->",item)
```

### using enumerate() works with any iterable while range(len()) only works with countable, indexable objects.

# True False

```python
if attr:
    print ('attr is truthy!')

if not attr:
    print ('attr is falsey!')

if attr is None:
    print ('attr is None!')
```

```python
# pseudoesque
a = [3, 4, 5]
b = []
for i in a:
    if i > 4:
        b.append(i)
```

```python
# pythonic
a = [3, 4, 5]
b = [i for i in a if i > 4]
# Or:
b = filter(lambda x: x > 4, a)
```

```python
a = [3, 4, 5]
for i in range(len(a)):
    a[i] += 3
```

```python
a = [3, 4, 5]
a = [i + 3 for i in a]
# Or:
a = map(lambda i: i + 3, a)
```
